```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Lets take a non-linear dataset from sklearn

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
x
```

```python
y
```

- x data as two column.
- y data has one column.

```python
data = {
    'column1':x[:,0],
    'column2':x[:,1],
    'output':y
}

df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
sns.scatterplot(x='column1', y='column2', data=df)
plt.show()
```

### Now lets use DBSCAN.

```python
y_data = df['output']
x_data = df.drop(columns=['output'])
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2, min_samples=5)
```

```python
pred = db.fit_predict(x_data)
```

```python
pred_data = x_data.copy()
pred_data['prediction'] = pred
```

### Now lets check original grouping and predicted clusters

```python
sns.scatterplot(x='column1', y='column2', hue='output', data=df)
plt.show()
```

```python
sns.scatterplot(x='column1', y='column2', hue='prediction', data=pred_data)
plt.show()
```

- The predicted clustering is so accurate.
