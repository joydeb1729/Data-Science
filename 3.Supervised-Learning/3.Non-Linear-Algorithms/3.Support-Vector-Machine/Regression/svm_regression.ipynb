```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
placement_package_data = pd.read_csv('placement_package.csv')
placement_package_data.head(3)
```

```python
placement_package_data.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa', y='package', data=placement_package_data)
plt.show()
```

```python
x = placement_package_data[['cgpa']]
y = placement_package_data['package']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
```

### 1.Linear Kernel

```python
svr = SVR(kernel='linear')
svr.fit(x_train, y_train)
svr.score(x_test, y_test)*100, svr.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='package', data=placement_package_data)
plt.scatter(x, svr.predict(x), c='red')
plt.title('Linear Data')
plt.show()
```

### 2.Polynomial Kernel

```python
svr = SVR(kernel='poly')
svr.fit(x_train, y_train)
svr.score(x_test, y_test)*100, svr.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='package', data=placement_package_data)
plt.scatter(x, svr.predict(x), c='red')
plt.title('Linear Data')
plt.show()
```

### 2.RBF Kernel

```python
svr = SVR(kernel='rbf')
svr.fit(x_train, y_train)
svr.score(x_test, y_test)*100, svr.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='package', data=placement_package_data)
plt.scatter(x, svr.predict(x), c='red')
plt.title('Linear Data')
plt.show()
```

### 4.Sigmoid Kernel

```python
svr = SVR(kernel='sigmoid')
svr.fit(x_train, y_train)
svr.score(x_test, y_test)*100, svr.score(x_train, y_train)*100
```
