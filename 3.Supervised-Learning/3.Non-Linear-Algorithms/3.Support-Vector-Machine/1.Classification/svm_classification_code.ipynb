```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
placement_data = pd.read_csv('placement.csv')
placement_data.head(3)
```

```python
placement_data.isnull().sum()
```

```python
sns.scatterplot(x='cgpa', y='score', hue='placed', data=placement_data)
plt.show()
```

```python
polynomial_data = pd.read_csv('polynomial_input.csv')
polynomial_data.head(3)
```

```python
polynomial_data.isnull().sum()
```

```python
sns.scatterplot(x='data1', y='data2', hue='output', data=polynomial_data)
plt.show()
```

```python
x = placement_data.iloc[:,:-1]
y = placement_data['placed']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
px = polynomial_data.iloc[:,:-1]
py = polynomial_data['output']
```

```python
px_train, px_test, py_train, py_test = train_test_split(px, py, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

### 1.Linear Kernel

```python
svc = SVC(kernel='linear')
svc.fit(x_train, y_train)

svc.score(x_test, y_test)*100 
```

```python
svc.score(x_test, y_test)*100 , svc.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc)
plt.title("Placement Data")
plt.show()
```

```python
psvc = SVC(kernel='linear')
psvc.fit(px_train, py_train)

psvc.score(px_test, py_test)*100, psvc.score(px_train, py_train)*100
```

```python
plot_decision_regions(px.to_numpy(), py.to_numpy(), clf=psvc)
plt.title("Non Linear Data")
plt.show()
```

```python
svc = SVC(kernel='poly')
svc.fit(x_train, y_train)

svc.score(x_test, y_test)*100 , svc.score(x_train, y_train)*100
```
