```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
placement_data = pd.read_csv('placement.csv')
placement_data.head()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa', y='score', hue='placed', data=placement_data)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=placement_data['cgpa'])
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=placement_data['score'])
plt.show()
```

```python
x = placement_data.iloc[:,:-1]
y = placement_data['placed']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
gnb.score(x_test, y_test)*100, gnb.score(x,y)*100
```

### Using **Gaussian Naive Bayes**

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

### Using **Multinomial Naive Bayes**

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
mnb.score(x_test, y_test)*100, mnb.score(x,y)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

### Using **Bernoulli Naive Bayes**

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
bnb.score(x_test, y_test)*100, bnb.score(x,y)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show()
```
